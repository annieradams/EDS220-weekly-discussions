####  How many hurricanes with Saffir-Simpson category 5 have been registered and what was their duration?

```python
df.dtypes
```

```python
df.loc[:, "start.date"] = pd.to_datetime(df.loc[:, 'start.date'])
df.loc[:, 'end.date'] = pd.to_datetime(df.loc[:, 'end.date'])

df_cat5 = df.loc[df['ss'] ==5, ].copy()


df_cat5.loc[:,'duration'] = df_cat5.loc[:,'end.date'] - df_cat5.loc[:,'start.date']
print("There have been ", len(df_cat5), " catagory 5 hurricanes")
```

####  Visualize Saffir-Simpson categories across time

```python
fig, ax = plt.subplots(figsize=(12, 5))

plt.scatter(df['start.date'], df['ss'],
           c='#3492eb')

plt.yticks([1,2,3,4,5])

plt.xlabel("Start Date")
plt.ylabel("Hurricane Category")

plt.title("Yucatan Peninsula: Hurricanes Over Time")

plt.show()
```
