#### Load libraries and hurricane data

```python
import pandas as pd
import matplotlib.pyplot as plt

# Importing data
df = pd.read_csv('../data/hf071-01-hurricanes.csv')
```

#### Question: How many hurricanes with Saffir-Simpson category 5 have been registered and what was their duration?

##### Data Wrangling

```python
(df.loc[df['ss'] == 5]
    .assign(duration = pd.to_datetime(df['end.date']) - pd.to_datetime(df['start.date'])))[['name','ss','duration']]
```

##### BONUS: Visualize Saffir-Simpson categories across time

```python
df['start.date'] = pd.to_datetime(df['start.date'])
df['end.date'] = pd.to_datetime(df['end.date'])
```

```python
plt.figure()
plt.scatter(pd.to_datetime(df['start.date']), df['ss'])

plt.title("Start date versus Saffir Scale")
plt.xlabel('Start date')
plt.ylabel('Saffir Scale')


plt.yticks([1, 2, 3, 4, 5])

plt.show()
```

```python
fig, ax = plt.subplots()

df.plot(ax=ax,
        x='start.date',
        y='ss',
        kind='scatter', 
        alpha=0.5 )

ax.set_title('')
ax.set_yticks([1,2,3,4,5])  # note that we don't need to specify labels
ax.set_ylabel('Saffir-Simpson Category')
ax.set_xlabel(' ')

ax.set_title('Saffir-Simpson Category of Hurricanes in the Yucatan Peninsula (1851-2000)')

plt.show
```
